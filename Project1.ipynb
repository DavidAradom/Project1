```python
# filtered to 10 states: top 5 states (CA, FL, TX, GA, OH) and bottom states (NH, ME, RI, AK, VT)
ten_state_crime = crime_df[["California", "Florida", "Texas", "Georgia", "Ohio", "New Hampshire", "Maine", "Rhode Island", "Arkansas", "Vermont"]]

# calculate avg crime count for years 1970 - 2014 for each state
ca_avg_rate = ten_state_crime["California"].loc[10:54].mean()
fl_avg_rate = ten_state_crime["Florida"].loc[10:54].mean()
tx_avg_rate = ten_state_crime["Texas"].loc[10:54].mean()
ga_avg_rate = ten_state_crime["Georgia"].loc[10:54].mean()
oh_avg_rate = ten_state_crime["Ohio"].loc[10:54].mean()
nh_avg_rate = ten_state_crime["New Hampshire"].loc[10:54].mean()
me_avg_rate = ten_state_crime["Maine"].loc[10:54].mean()
ri_avg_rate = ten_state_crime["Rhode Island"].loc[10:54].mean()
ak_avg_rate = ten_state_crime["Arkansas"].loc[10:54].mean()
vt_avg_rate = ten_state_crime["Vermont"].loc[10:54].mean()

```

```python
crime_summary_table = pd.DataFrame({"California": ca_avg_rate,
                                    "Florida": [fl_avg_rate],
                                    "Texas": [tx_avg_rate],
                                    "Georgia": [ga_avg_rate],
                                    "Ohio": [oh_avg_rate],
                                    "New Hampshire": [nh_avg_rate],
                                    "Maine": [me_avg_rate],
                                    "Rhode Island": [ri_avg_rate],
                                    "Arkansas": [ak_avg_rate],
                                    "Vermont": [vt_avg_rate]})

crime_summary_table
```
